```python
def find_students_above_threshold(students_data, threshold):
    above_threshold_students = {}
    for student, score in students_data.items():
        if score >= threshold:
            above_threshold_students[student] = score
    return above_threshold_students

def find_failures(students_data):
    return find_students_above_threshold(students_data, 35)

def find_passing_students(students_data):
    return find_students_above_threshold(students_data, 35)

def find_top_scorers(students_data):
    top_scorers = {}
    for student, score in students_data.items():
        if score > 90:
            top_scorers[student] = score
    return top_scorers

def main():
    # Create a dictionary with students' names and their marks
    students_marks = {
        "Alice": 75,
        "Bob": 60,
        "Charlie": 45,
        "David": 25,
        "Eva": 80,
        "Fiona": 95,
        "George": 88
    }

    # Find students who failed to get thirty-five
    failures = find_failures(students_marks)

    # Find students who got pass marks
    passing_students = find_passing_students(students_marks)

    # Find top scorers (students with more than 90 percent)
    top_scorers = find_top_scorers(students_marks)

    # Print the solutions
    print("Students who failed to get thirty-five:")
    for student in failures.keys():
        print(student)

    print("\nStudents who got pass marks:")
    for student in passing_students.keys():
        print(student)

    print("\nTop Scorers (students with more than 90 percent):")
    for student, score in top_scorers.items():
        print(f"{student}: {score}%")

if __name__ == "__main__":
    main()
```

```python
# Constants for designation access levels
VIEWER = 1
EDITOR = 2
CREATOR = 3

def get_user_designation():
    # Simulate getting the user's designation, replace this with actual logic in a real system
    # For this example, we'll use input for simplicity
    while True:
        try:
            user_input = int(input("Enter your designation (1 for Viewer, 2 for Editor, 3 for Creator): "))
            if user_input in (VIEWER, EDITOR, CREATOR):
                return user_input
            else:
                print("Invalid input. Please enter 1, 2, or 3.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

def can_view(user_designation):
    return user_designation >= VIEWER

def can_edit(user_designation):
    return user_designation >= EDITOR

def can_create(user_designation):
    return user_designation >= CREATOR

def main():
    user_designation = get_user_designation()

    if can_create(user_designation):
        print("You can view, edit, and create the file.")
    elif can_edit(user_designation):
        print("You can view and edit the file.")
    elif can_view(user_designation):
        print("You can view the file.")
    else:
        print("Sorry, you don't have permission to access the file.")

if __name__ == "__main__":
    main()
```
